```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv('/content/Churn.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
df["Exited"] = df["Exited"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Exited', axis=1), df['Exited'], test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

object_columns = X_train.select_dtypes(include=['object']).columns
numeric_columns = X_train.select_dtypes(include=['float64', 'int64']).columns

num_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

object_pipeline = Pipeline([
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

transformer = ColumnTransformer([
        ('num', num_pipeline, numeric_columns),
        ('cat', object_pipeline, object_columns)]
)

X_train_transformed = transformer.fit_transform(X_train)
X_test_transformed = transformer.transform(X_test)

feature_names = transformer.get_feature_names_out()

X_train_transformed = pd.DataFrame(X_train_transformed, columns = feature_names)
X_test_transformed = pd.DataFrame(X_test_transformed, columns = feature_names)

X_train_transformed
```

```python
X_train_transformed.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=[21]),
    tf.keras.layers.Dense(300, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train_transformed, y_train, epochs=10, validation_data=(X_test_transformed, y_test))
```
