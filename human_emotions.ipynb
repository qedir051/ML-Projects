```python
import os
import glob

# Specify the root directory where images and folders are located
root_dir = '/content/6 Emotions for image classification'

# Loop through each subdirectory and remove non-JPG files
for folder in glob.glob(os.path.join(root_dir, '*')):
    if os.path.isdir(folder):  # Ensure it's a folder
        for file in glob.glob(os.path.join(folder, '*')):
            # Check if the file does not have .jpg or .jpeg extension (case insensitive)
            if not (file.lower().endswith('.jpg') or file.lower().endswith('.jpeg')):
                try:
                    os.remove(file)  # Delete the file
                    print(f"Deleted: {file}")
                except Exception as e:
                    print(f"Error deleting {file}: {e}")
```

```python
import os
import glob
from PIL import Image

# Specify the root directory where images and folders are located
root_dir = '/content/6 Emotions for image classification'

# Loop through each subdirectory and process .jpg and .jpeg files
for folder in glob.glob(os.path.join(root_dir, '*')):
    if os.path.isdir(folder):  # Ensure it's a folder
        for file_path in glob.glob(os.path.join(folder, '*')):
            if file_path.lower().endswith(('.jpg', '.jpeg')):  # Process both .jpg and .jpeg
                try:
                    # Open the image
                    img = Image.open(file_path)

                    # Convert image to RGB if needed
                    img = img.convert('RGB')

                    # Save the image as .jpeg by changing the extension
                    new_file_path = os.path.splitext(file_path)[0] + '.jpeg'
                    img.save(new_file_path, 'JPEG')

                    # Optionally, remove the original file if it was .jpg
                    if file_path.lower().endswith('.jpg'):
                        os.remove(file_path)
                        print(f"Converted and removed: {file_path}")

                    print(f"Converted: {file_path} to {new_file_path}")

                except Exception as e:
                    print(f"Error processing {file_path}: {e}")
```

```python
IMG_SIZE = (224,224)
BATCH_SIZE = 32

data_dir = '/content/6 Emotions for image classification'

# train_data, validation_data = tf.keras.utils.image_dataset_from_directory(data_dir,
#                                                                  image_size=IMG_SIZE,
#                                                                  batch_size=BATCH_SIZE,
#                                                                  label_mode='categorical',
#                                                                  subset='both',
#                                                                  validation_split=0.2,
#                                                                  seed=42)

train_data = tf.keras.utils.image_dataset_from_directory(data_dir,
                                                                 image_size=IMG_SIZE,
                                                                 batch_size=BATCH_SIZE,
                                                                 label_mode='categorical',
                                                                 subset='training',
                                                                 validation_split=0.2,
                                                                 seed=42)

validation_data = tf.keras.utils.image_dataset_from_directory(data_dir,
                                                                 image_size=IMG_SIZE,
                                                                 batch_size=BATCH_SIZE,
                                                                 label_mode='categorical',
                                                                 subset='validation',
                                                                 validation_split=0.2,
                                                                 seed=42)
```

```python
train_data.class_names
```

```python
activation = 'relu'
kernel_initializer = 'he_normal'
base_model = Sequential([
    Input(shape=(224,224,3)),
    tf.keras.layers.Rescaling(1./255),
    Conv2D(32, 2, activation=activation, kernel_initializer=kernel_initializer),
    Conv2D(32, 2, activation=activation, kernel_initializer=kernel_initializer),
    MaxPooling2D(2),
    Conv2D(64, 2, activation=activation, kernel_initializer=kernel_initializer),
    Conv2D(64, 2, activation=activation, kernel_initializer=kernel_initializer),
    MaxPooling2D(2),
    Flatten(),
    Dense(6, activation="softmax")
])
```

```python
base_model.compile(loss="categorical_crossentropy",
                   optimizer = tf.keras.optimizers.Adam(),
                   metrics=['accuracy'])
```

```python
base_model.fit(train_data, epochs=5, validation_data= validation_data)
```

```python
base_model = tf.keras.applications.EfficientNetV2S(include_top=False)

base_model.trainable = False

inputs = tf.keras.layers.Input(shape = (224,224,3))

x = base_model(inputs)
print(f"shape after inputs through base model: {x.shape}")

x = tf.keras.layers.GlobalAveragePooling2D(name = "Global_average_pooling_layer")(x)
print(f"Shape after globalAveragePoolingLayer {x.shape}")

outputs = tf.keras.layers.Dense(6 ,activation="softmax", name="output_layer")(x)

model = tf.keras.Model(inputs, outputs)

model.compile(loss="categorical_crossentropy",
                optimizer=tf.keras.optimizers.Amodam(0.003),
                metrics=["accuracy"])

history_model = model.fit(train_data,
            epochs=20,
            batch_size = BATCH_SIZE,
            validation_data=validation_data)
```
